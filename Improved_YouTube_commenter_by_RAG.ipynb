```python

from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.core import Settings, SimpleDirectoryReader, VectorStoreIndex
from llama_index.core.retrievers import VectorIndexRetriever
from llama_index.core.query_engine import RetrieverQueryEngine
from llama_index.core.postprocessor import SimilarityPostprocessor
```

Define settings

```python

# import any embedding model on HF hub (https://huggingface.co/spaces/mteb/leaderboard)
Settings.embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-small-en-v1.5")
# Settings.embed_model = HuggingFaceEmbedding(model_name="thenlper/gte-large") # alternative model

Settings.llm = None
Settings.chunk_size = 256
Settings.chunk_overlap = 25
```

```python
import shutil
import os

# Define paths
pdf_paths = [
    "/content/4 Ways to Quantify Fat Tails with Python _ by Shaw Talebi _ Towards Data Science.pdf",
    "/content/Detecting Power Laws in Real-world Data with Python _ by Shawhin Talebi _ Towards Data Science.pdf",
    "/content/Pareto, Power Laws, and Fat Tails _ by Shaw Talebi _ Towards Data Science _ Towards Data Science.pdf"
]

# Directory to move PDFs to
dest_dir = "/content/articles"

# Create the directory if it does not exist
if not os.path.exists(dest_dir):
    os.makedirs(dest_dir)

# Move PDFs to the directory
for pdf_path in pdf_paths:
    shutil.move(pdf_path, os.path.join(dest_dir, os.path.basename(pdf_path)))

print("PDFs moved to the 'articles' directory.")
```

```python
# articles available here: {add GitHub repo}
documents = SimpleDirectoryReader("articles").load_data()
```

```python

# some ad hoc document refinement
print(len(documents))
for doc in documents:
    if "Member-only story" in doc.text:
        documents.remove(doc)
        continue

    if "The Data Entrepreneurs" in doc.text:
        documents.remove(doc)

    if " min read" in doc.text:
        documents.remove(doc)

print(len(documents))
```

```python

# store docs into vector DB
index = VectorStoreIndex.from_documents(documents)
```

Set up search function

```python

# set number of docs to retreive
top_k = 3

# configure retriever
retriever = VectorIndexRetriever(
    index=index,
    similarity_top_k=top_k,
)
```

```python
# assemble query engine
query_engine = RetrieverQueryEngine(
    retriever=retriever,
    node_postprocessors=[SimilarityPostprocessor(similarity_cutoff=0.5)],
)
```

Retrieve relevant docs

```python

# query documents
query = "What is fat-tailedness?"
response = query_engine.query(query)
```

```python

# reformat response
context = "Context:\n"
for i in range(top_k):
    context = context + response.source_nodes[i].text + "\n\n"

print(context)
```

Import LLMs

```python
from peft import PeftModel, PeftConfig
from transformers import AutoModelForCausalLM, AutoTokenizer
from transformers import GPTQConfig

model_name = "TheBloke/Mistral-7B-Instruct-v0.2-GPTQ"

# Create a GPTQConfig object
quantization_config = GPTQConfig(bits=4, disable_exllama=True)

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto",
    trust_remote_code=False,
    revision="main",
    quantization_config=quantization_config
)

config = PeftConfig.from_pretrained("shawhin/shawgpt-ft")
model = PeftModel.from_pretrained(model, "shawhin/shawgpt-ft")

# load tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=True)
```

```python

# prompt (no context)
intstructions_string = f"""SatvikGPT, functioning as a virtual data science consultant on YouTube, communicates in clear, accessible language, escalating to technical depth upon request. \
It reacts to feedback aptly and ends responses with its signature '–SatvikGPT'. \
SatvikGPT will tailor the length of its responses to match the viewer's comment, providing concise acknowledgments to brief expressions of gratitude or feedback, \
thus keeping the interaction natural and engaging.

Please respond to the following comment.
"""
prompt_template = lambda comment: f'''[INST] {intstructions_string} \n{comment} \n[/INST]'''
```

```python
comment = "What is fat-tailedness?"

prompt = prompt_template(comment)
print(prompt)
```

```python
model.eval()

inputs = tokenizer(prompt, return_tensors="pt")
outputs = model.generate(input_ids=inputs["input_ids"].to("cuda"), max_new_tokens=280)

print(tokenizer.batch_decode(outputs)[0])
```

```python
# prompt (with context)
prompt_template_w_context = lambda context, comment: f"""[INST]SatvikGPT, functioning as a virtual data science consultant on YouTube, communicates in clear, accessible language, escalating to technical depth upon request. \
It reacts to feedback aptly and ends responses with its signature '–SatvikGPT'. \
SatvikGPT will tailor the length of its responses to match the viewer's comment, providing concise acknowledgments to brief expressions of gratitude or feedback, \
thus keeping the interaction natural and engaging.

{context}
Please respond to the following comment. Use the context above if it is helpful.

{comment}
[/INST]
"""
```

```python
prompt = prompt_template_w_context(context, comment)

inputs = tokenizer(prompt, return_tensors="pt")
outputs = model.generate(input_ids=inputs["input_ids"].to("cuda"), max_new_tokens=280)

print(tokenizer.batch_decode(outputs)[0])
```
